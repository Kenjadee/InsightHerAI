```python
df = pd.read_csv('Maternal Health Risk Data Set.csv')
print(df.head())
```

```python
mapping = {'low risk': 0, 'mid risk': 1, 'high risk': 2}
df['RiskLevel'] = df['RiskLevel'].map(mapping)
```

```python
x = df.drop('RiskLevel', axis= 1)
y = df['RiskLevel']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
print(x_train, x_test, y_train, y_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler_standard = StandardScaler()
x_train_scaled = scaler_standard.fit_transform(x_train)
x_test_scaled = scaler_standard.transform(x_test)
```

```python
y_train.value_counts(normalize=True)
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(max_iter= 5000)
lr_model.fit(x_train_scaled, y_train)
```

```python
lr_predictions = lr_model.predict(x_test_scaled)
```

```python
print("Predictions (first 20):")
print(lr_predictions[:20])

print("\nActual values (first 20):")
print(y_test.values[:20])

from sklearn.metrics import accuracy_score
lr_accuracy = accuracy_score(y_test, lr_predictions)
print(lr_accuracy)

from sklearn.metrics import classification_report
cr = classification_report(y_test, lr_predictions, target_names=["GREEN", "YELLOW", "RED"])
print(cr)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(lr_model, x, y, cv=5)
print(scores)
print(scores.mean())

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, lr_predictions)
print (cm)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state=42)

dt_model.fit(x_train_scaled, y_train)

dt_predictions = dt_model.predict(x_test_scaled)
print("Predictions (first 20):")
print(dt_predictions[:20])

print("\nActual values (first 20):")
print(y_test.values[:20])

from sklearn.metrics import accuracy_score
dt_accuracy = accuracy_score(y_test, dt_predictions)
print(dt_accuracy)

from sklearn.metrics import classification_report
cr = classification_report(y_test, dt_predictions, target_names=["GREEN", "YELLOW", "RED"])
print(cr)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(dt_model, x, y, cv=5)
print(scores)
print(scores.mean())

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, dt_predictions)
print (cm)

```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(x_train_scaled, y_train)
rf_predictions = rf_model.predict(x_test_scaled)
dt_predictions = dt_model.predict(x_test_scaled)

print("Predictions (first 20):")
print(rf_predictions[:20])

print("\nActual values (first 20):")
print(y_test.values[:20])

rf_accuracy = accuracy_score(y_test, rf_predictions)
print(rf_accuracy)
cr = classification_report(y_test, rf_predictions, target_names=["LOW", "MID", "HIGH"])
print(cr)

from sklearn.model_selection import cross_val_score

scores = cross_val_score(rf_model, x, y, cv=5)
print(scores)
print(scores.mean())

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, rf_predictions)
print (cm)
```

```python
from sklearn.metrics import roc_auc_score
y_prob = rf_model.predict_proba(x_test_scaled)

auc_score = roc_auc_score(
    y_test,
    y_prob,
    multi_class='ovr',
    average='macro'
)

print("ROC-AUC Score:", auc_score)
```

```python
import pandas as pd

feature_importance = pd.DataFrame({
    'Feature': x.columns,
    'Importance': rf_model.feature_importances_
}).sort_values('Importance', ascending=False)

print(feature_importance)
```

```python
def get_confidence(probabilities):
  max_prob = max(probabilities)

  if max_prob >= 0.8:
    return "HIGH CONFIDENCE"
  elif max_prob >= 0.6:
    return "LOW CONFIDENCE - SUGGEST REVIEW"
  else:
    return "UNCERTAIN - ESCALATE TO SENIOR WORKER"
```

```python
print(x.columns.tolist())
```

```python
feature_names = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']

def predict_triage(patient_vitals):
    vitals_df = pd.DataFrame([patient_vitals], columns=feature_names)
    vitals_scaled = scaler_standard.transform(vitals_df)
    probabilities = rf_model.predict_proba(vitals_scaled)[0]
    predicted_class = rf_model.predict(vitals_scaled)[0]
    confidence = get_confidence(probabilities)

    triage_map = {
        0: "Low Risk",
        1: "Mid Risk",
        2: "High Risk"
    }

    decision = triage_map[predicted_class]

    print("\n   INSIGHTHER ASSESSMENT")
    print("-----------------------------------")
    print(f"DECISION:    {decision}")
    print(f"CONFIDENCE:  {confidence}")
    print("--------------------------------")
    print(f"Low Risk:       {probabilities[0]:.1%}")
    print(f"Mid Risk:      {probabilities[1]:.1%}")
    print(f"High Risk:         {probabilities[2]:.1%}")
    print("--------------------------------\n")

    return {
        "triage": decision,
        "confidence": confidence,
        "probabilities": {
            "Low Risk": round(probabilities[0], 3),
            "Mid Risk": round(probabilities[1], 3),
            "High Risk": round(probabilities[2], 3)
        }
    }
```

```python
patient_1 = [
    13, 99, 70, 20, 50, 60
]

predict_triage(patient_1)
```
